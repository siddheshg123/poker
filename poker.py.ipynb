```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Aug  1 19:33:44 2018

@author: kpit
"""



import itertools, random
import pandas as pd
import csv

df = pd.read_csv('pokeruser.csv',sep=',',header=0) 
df=df.set_index('USERNAME')
df
user=input("user_name:")
if user not in df.index:
    print("invalid user name")
else:
    password=df.at[user,'PASSWORD']
    if input('Password:')==password:
        df.at[user,'COUNT']+=1
        df.to_csv("pokeruser.csv")
        print("play")
        # make a deck of cards
deck = list(itertools.product(range(1,14),['Spade','Heart','Diamond','Club']))
print (deck)
cards_player1=[]
cards_player2=[]
random.shuffle(deck)

# draw five cards
print("player1 got:")
for i in range(3):
    cards_player1.append(deck[i])
    print(deck[i][0], "of", deck[i][1])
print("\n")   
print("\n")
print("player2 got:")
for i in range(3):
    cards_player2.append(deck[len(deck)-1-i])
    print(deck[len(deck)-1-i][0], "of", deck[len(deck)-1-i][1])
print("\n")   
print("\n")
print("cards on board:")
for i in range(2):
    cards_player1.append(deck[i+10])
    cards_player2.append(deck[i+10])
    print(deck[i+10][0], "of", deck[i+10][1])
    
print(cards_player2) 
print(cards_player1)
def roy(ls):
    num=0
    for i in range(len(ls)):
        num=num+ls[i][0]
    print(num)
    
roy(cards_player1)
roy(cards_player2)

def royal_flush(ls):
    card_value=[]
    card_suit=[]
    for i in range(len(ls)):
        card_value.append(ls[i][0])
        card_suit.append(ls[i][1])
    num=card_suit.count(card_suit[0])
    if num==5:
        low = min(card_value)
        if low == 10 and (low + 1 in card_value) and (low + 2 in card_value) and (low + 3 in card_value) and (low + 4 in card_value):
            return True
        else:
            return False
    else:
        return False
            
def flush(ls):
    card_suit = []
    for i in range(len(ls)):
        card_suit.append(ls[i][1])
    current_suit=card_suit[0]
    for card in card_suit:
        if card!= current_suit:
            return False
        current_suit=card
    return True           
                     
def four_of_a_kind(ls):
    card_value=[]
    for i in range(len(ls)):
        card_value.append(ls[i][0])
    count_first_card=card_value.count(card_value[0])
    count_second_card=card_value.count(card_value[1])
    if count_first_card==4 | count_second_card==4:
         return True
    else:
         return False
def three_of_a_kind(ls):
    card_value=[]
    for i in range(len(ls)):
        card_value.append(ls[i][0])
    count_first_card=card_value.count(card_value[0])
    count_second_card=card_value.count(card_value[1])
    if count_first_card==3 | count_second_card==3:
         return True
    else:
         return False
         
def straight(ls):
    card_value=[]
    for i in range(len(ls)):
        card_value.append(ls[i][0])
    low=min(card_value)
    if low == 10 and (low + 1 in card_value) and (low + 2 in card_value) and (low + 3 in card_value) and (low + 4 in card_value):
         return True
    else:
         return False
def fullhouse(ls):
        card_value = []
        for i in range(len(ls)):
            card_value.append(ls[i][0])

        card_value.sort()
        cardOneCount = card_value.count(card_value[0])
        cardTwoCount = card_value.count(card_value[len(card_value) - 1])

        if (cardOneCount == 2 and cardTwoCount == 3 ) or (cardOneCount == 3 and cardTwoCount == 2):
            return True
        else: 
            return False            
def two_pair(ls):
    card_value=[]
    for i in range(len(ls)):
        card_value.append(ls[i][0])
        
    cardOneCount = card_value.count(card_value[0])
    cardTwoCount = card_value.count(card_value[4])
    cardThreeCount = card_value.count(card_value[2])

    if(cardOneCount == 2 and cardTwoCount == 2) or (cardOneCount == 2 and cardThreeCount == 2) | (cardTwoCount == 2 and cardThreeCount == 2):
        return True
    return False
def getScore(ls):
    score = 0
    if royal_flush(ls): score = 140000
    elif flush(ls) and straight(ls): score = 130000
    elif four_of_a_kind(ls): score = 120000
    elif fullhouse(ls): score = 110000
    elif flush(ls): score = 100000
    elif straight(ls): score = 90000
    elif three_of_a_kind(ls): score = 80000
    elif two_pair(ls): score = 70000
    else: score = 10000
    return score

player_1_score=getScore(cards_player1)
player_2_score=getScore(cards_player2)

print(player_1_score)
print(player_2_score)


    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    



        



    
    

```
